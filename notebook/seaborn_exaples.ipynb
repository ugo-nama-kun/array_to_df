# Replot Example
Concatinating multiple plot lines into one DataFrame

```python
# data
time_tick = np.arange(20)
result1 = 1 * np.random.rand(len(time_tick))
result2 = 2 * np.random.rand(len(time_tick))
result3 = 3 * np.random.rand(len(time_tick))

# dataframes. "label" is important!!!
df1 = pd.DataFrame({"x": time_tick,
                    "y": result1,
                    "label": [1]*len(result1)})

df2 = pd.DataFrame({"x": time_tick,
                    "y": result2,
                    "label": [2]*len(result2)})

df3 = pd.DataFrame({"x": time_tick,
                    "y": result3,
                    "label": [3]*len(result3)})

print(df1, df2, df3)

# This process concatinate df
df = pd.concat([df1, df2, df3])

print(df)
```

```python
sns.relplot(
    x="x",
    y="y",
    hue="label",
    data=df,
    kind="line"
)
```

```python
sns.relplot(
    x="x",
    y="y",
    col="label",
    data=df,
    kind="line"
)
```

```python
sns.relplot(
    x="x",
    y="y",
    row="label",
    data=df,
    kind="line"
)
```

```python
sns.relplot(
    x="x",
    y="y",
    hue="label",
    data=df,
    kind="line",
    aspect=2
)
```

```python
sns.relplot(
    x="x",
    y="y",
    hue="label",
    data=df,
    kind="scatter",
    aspect=2
)
```

```python
# 95% confidence interval
sns.relplot(x="x",
            y="y",
            kind="line",
            data=df);
```

```python
# Standard deviation
sns.relplot(x="x",
            y="y",
            kind="line",
            ci="sd",
            data=df);
```

```python
# just mean plot
sns.relplot(x="x",
            y="y",
            kind="line",
            ci=None,
            data=df);
```

```python
time_tick = np.arange(5)
result1 = 1 * np.random.rand(len(time_tick))
result2 = 2 * np.random.rand(len(time_tick))
result3 = 3 * np.random.rand(len(time_tick))

# dataframes. "label" is important!!!
df1 = pd.DataFrame({"x": time_tick,
                    "y": result1})

df2 = pd.DataFrame({"x": time_tick,
                    "y": result2})

df3 = pd.DataFrame({"x": time_tick,
                    "y": result3})

# This process concatinate df
df = pd.concat([df1, df2, df3], keys=["a", "b", "c"])
print(df)
sns.relplot(x="x",
            y="y",
            kind="line",
            data=df);
```

# Bar plot for Categorical data
Control of figure aesthetics
https://seaborn.pydata.org/tutorial/aesthetics.html?highlight=styles

Categorical data plot
https://seaborn.pydata.org/tutorial/categorical.html

```python
sns.set_theme(style="ticks", color_codes=True)

categories = ["red", "blue", "green"]

n = 20
hist_red = np.random.randn(n) + 1
hist_blue = np.random.randn(n) + 2
hist_green = np.random.randn(n) + 3

df_red = pd.DataFrame(
    {
        "y" : hist_red,
        "label": ["red"] * len(hist_red)
    }
)

df_blue = pd.DataFrame(
    {
        "y" : hist_blue,
        "label": ["blue"] * len(hist_blue)
    }
)

df_green = pd.DataFrame(
    {
        "y" : hist_green,
        "label": ["green"] * len(hist_green)
    }
)

df = pd.DataFrame()
df = pd.concat([df, df_red, df_blue, df_green])
print(df)

sns.catplot(x="label", y="y", data=df)
```

```python
sns.catplot(x="label", y="y", jitter=False, data=df)
```

```python
sns.catplot(x="label", y="y", kind="box", data=df)
```

```python
sns.catplot(x="label", y="y", kind="violin", data=df)
```

```python
n = 20

df_list = []
df = pd.DataFrame()
for experiment in ["he", "she", "they"]:
    hist_red = np.random.randn(n) + 1
    hist_blue = np.random.randn(n) + 2
    hist_green = np.random.randn(n) + 3

    df_red = pd.DataFrame(
        {
            "y" : hist_red,
            "label": ["method 1"] * len(hist_red),
            "experiment": [experiment] * len(hist_red)
        }
    )

    df_blue = pd.DataFrame(
        {
            "y" : hist_blue,
            "label": ["my way"] * len(hist_blue),
            "experiment": [experiment] * len(hist_blue)
        }
    )

    df_green = pd.DataFrame(
        {
            "y" : hist_green,
            "label": ["awesome idea"] * len(hist_green),
            "experiment": [experiment] * len(hist_green)
        }
    )
    
    df = pd.concat([df, df_red, df_blue, df_green])

df
```

```python
sns.catplot(x="label",
            y="y",
            kind="bar",
            hue="experiment",
            data=df)
```

```python
sns.catplot(x="label", y="y", kind="point", hue="experiment", data=df)
```
